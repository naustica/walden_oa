```python
import holoviews as hv
from holoviews import opts
from google.cloud import bigquery
import pandas as pd
from bokeh.resources import INLINE
import panel as pn
```

```python
hv.extension('bokeh')
```

```python
client = bigquery.Client(project='subugoe-collaborative')
```

```python
df_diff = client.query(f"""
                    SELECT COUNT(DISTINCT(doi)) AS n
                    FROM (
                      SELECT doi,
                            original_oa_status,
                            CASE
                                WHEN best_oa_location_is_oa IS NULL THEN 'closed'
                                WHEN best_oa_location_source_type='repository' THEN 'green'
                                WHEN best_oa_location_source_is_in_doaj=TRUE OR (best_oa_location_source_is_oa=TRUE AND apc_list_value=0) THEN 'diamond'
                                -- WHEN (best_oa_location_source_is_in_doaj=TRUE OR best_oa_location_source_is_oa=TRUE) AND apc_list_value=0 THEN 'diamond'
                                WHEN (best_oa_location_source_is_in_doaj=TRUE OR best_oa_location_source_is_oa=TRUE) THEN 'gold'
                                WHEN (best_oa_location_source_is_in_doaj=FALSE AND best_oa_location_source_is_oa=FALSE)
                                    AND best_oa_location_license IS NOT NULL THEN 'hybrid'
                                WHEN (best_oa_location_source_is_in_doaj=FALSE AND best_oa_location_source_is_oa=FALSE) 
                                    AND best_oa_location_license IS NULL THEN 'bronze'
                                ELSE 'None'
                                END
                            AS new_oa_status
                      FROM subugoe-collaborative.resources.walden_oa_articles_18_25
                    ) 
                    WHERE original_oa_status != new_oa_status
                  """).to_dataframe()
```

```python
df = client.query(f"""
                  SELECT COUNT(DISTINCT(doi)) AS n,
                         original_oa_status,
                         CASE
                             WHEN best_oa_location_is_oa IS NULL THEN 'closed '
                             WHEN best_oa_location_source_type='repository' THEN 'green '
                             WHEN best_oa_location_source_is_in_doaj=TRUE OR (best_oa_location_source_is_oa=TRUE AND apc_list_value=0) THEN 'diamond '
                             -- WHEN (best_oa_location_source_is_in_doaj=TRUE OR best_oa_location_source_is_oa=TRUE) AND apc_list_value=0 THEN 'diamond '
                             WHEN (best_oa_location_source_is_in_doaj=TRUE OR best_oa_location_source_is_oa=TRUE) THEN 'gold '
                             WHEN (best_oa_location_source_is_in_doaj=FALSE AND best_oa_location_source_is_oa=FALSE)
                                 AND best_oa_location_license IS NOT NULL THEN 'hybrid '
                             WHEN (best_oa_location_source_is_in_doaj=FALSE AND best_oa_location_source_is_oa=FALSE) 
                                 AND best_oa_location_license IS NULL THEN 'bronze '
                             ELSE NULL
                             END
                         AS new_oa_status,
                         CASE
                             WHEN original_oa_status = 'closed' THEN '#e5e5e5'
                             WHEN original_oa_status = 'green' THEN '#40a954'
                             WHEN original_oa_status = 'diamond' THEN '#446ace'
                             WHEN original_oa_status = 'gold' THEN '#e9b121'
                             WHEN original_oa_status = 'hybrid' THEN '#97bbf5ff'
                             WHEN original_oa_status = 'bronze' THEN '#fd725d'
                             ELSE 'black'
                             END
                         AS edge_color
                  FROM subugoe-collaborative.resources.walden_oa_articles_18_25
                  GROUP BY original_oa_status, new_oa_status, edge_color
                  ORDER BY original_oa_status, new_oa_status, edge_color DESC
                  """).to_dataframe()
```

```python
df = df[['original_oa_status', 'new_oa_status', 'n', 'edge_color']]
```

```python
df = df.fillna(value='None')
```

```python
df['original_oa_status'].replace({'bronze': 'Bronze OA', 
                                  'closed': 'Closed', 
                                  'diamond': 'Diamond OA', 
                                  'gold': 'Gold OA', 
                                  'green': 'Green OA', 
                                  'hybrid': 'Hybrid OA'}, 
                                 inplace=True)

df['new_oa_status'].replace({'bronze ': 'Bronze OA ', 
                             'closed ': 'Closed ', 
                             'diamond ': 'Diamond OA ', 
                             'gold ': 'Gold OA ', 
                             'green ': 'Green OA ', 
                             'hybrid ': 'Hybrid OA '}, 
                            inplace=True)
```

```python
df
```

```python
node_colors_dict={
        'None': 'black',
        'Closed ': '#e5e5e5',
        'Closed': '#e5e5e5',
        'Green OA ': '#40a954',
        'Green OA': '#40a954',
        'Diamond OA ': '#446ace',
        'Diamond OA': '#446ace',
        'Gold OA ': '#e9b121',
        'Gold OA': '#e9b121',
        'Hybrid OA ': '#97bbf5ff',
        'Hybrid OA': '#97bbf5ff',
        'Bronze OA ': '#fd725d',
        'Bronze OA': '#fd725d',
}

sankey = hv.Sankey(df)
sankey.opts(
    opts.Sankey(
        cmap=node_colors_dict,
        edge_color='edge_color',
        node_padding=7,
        #label_text_font_size='9pt',
        width=650, 
        height=700,
        toolbar=None,
        default_tools=[]
    )
)
```

```python
pn.pane.HoloViews(sankey).save('sankey', embed=True, resources=INLINE)
```
